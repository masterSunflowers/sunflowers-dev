`````python

def has_close_elements(numbers, threshold: float) -> bool:
    for idx, elem in enumerate(numbers):
        for idx2, elem2 in enumerate(numbers):
            if idx != idx2:                
                distance = abs(elem - elem2)                
                if distance < threshold:                    
                    return True    
                return False

has_close_elements([1.0, 1.0], 0.0)
`````
